```python
import pandas as pd
```

```python
domestic_data=pd.read_csv("cleaned_resources/new_dom_clean_data.csv")
domestic_data.head()
```

```python
national_average_data=domestic_data[domestic_data['market']=='National Average']
```

```python
wholesale_average_data=national_average_data[national_average_data['price_type']=='Wholesale']
```

```python
wholesale_average_data
```

```python
wholesale_average_date = wholesale_average_data.groupby(["country", "commodity"])
```

```python
min_df = wholesale_average_date[["price"]].min()
```

```python
# min_df.reset_index()
```

```python
max_df = wholesale_average_date[["price"]].max()
```

```python
max_df.reset_index()
```

```python
median_df = wholesale_average_date[["price"]].median()
```

```python
median_df.reset_index()
```

```python
merged_df = pd.concat([min_df, max_df, median_df], axis=1)
```

```python
merged_df.columns = ['min_price', 'max_price', 'median_price']
```

```python
merged_df
```

```python
merged_df.reset_index()
```
