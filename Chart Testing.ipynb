```python
median_Maize = Ethopia_Maize['price'].median()
median_Sorghum = Ethopia_Sorghum['price'].median()
median_Teff = Ethopia_Teff['price'].median()
median_Wheat= Ethopia_Wheat['price'].median()
```

```python
max_Maize = Ethopia_Maize['price'].max()
max_Sorghum = Ethopia_Sorghum['price'].max()
max_Teff = Ethopia_Teff['price'].max()
max_Wheat= Ethopia_Wheat['price'].max()
print(max_Maize)
print(max_Sorghum)
print(max_Teff)
print(max_Wheat)
```

```python
min_Maize = Ethopia_Maize['price'].min()
min_Sorghum = Ethopia_Sorghum['price'].min()
min_Teff = Ethopia_Teff['price'].min()
min_Wheat= Ethopia_Wheat['price'].min()
print(min_Maize)
print(min_Sorghum)
print(min_Teff)
print(min_Wheat)
```

```python
# Assuming you have a dataframe called 'df' with two columns: 'Group' and 'Values'
Ethopia_box_chart = pd.DataFrame({'Commodities': ['Maize', 'Maize', 'Maize', 'Sorghum', 'Sorghum', 'Sorghum','Teff','Teff','Teff','Wheat','Wheat','Wheat'], 'Values': [min_Maize, median_Maize, max_Maize, min_Sorghum, median_Sorghum,max_Sorghum, min_Teff, median_Teff,max_Teff,min_Wheat,median_Wheat,max_Wheat]})

# Create the horizontal box chart
Ethopia_box_chart.boxplot(column='Values', by='Commodities', vert=False)

# Set labels and title
plt.xlabel('Prices Dollars USD')
plt.ylabel('Commodities')
plt.title('Ethopian Commodity Price Ranges')

# Show the chart
plt.show()
```
